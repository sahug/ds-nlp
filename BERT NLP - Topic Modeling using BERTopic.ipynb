**Create Model**

To create a model using `BERTopic`, we need to load the tweets as a list and then pass it to the `fit_transform` method. This method will do the following: 

- Fit the model on the collection of tweets. 
- Generate topics.
- Return the tweets with the topics.

```python
%pip install -qq bertopic 
```

```python
from bertopic import BERTopic
```

```python
# create model 
model = BERTopic(verbose=True, nr_topics=20)
 
# convert to list 
train = train.text.to_list()
 
topics, probabilities = model.fit_transform(train)
```

**Select Top Topics**

```python
model.get_topic_freq().head(11)
```

**Select One Topic**

```python
model.get_topic(6)
```

**Topic Modeling Visualization**

```python
model.visualize_heatmap()
```

**Topic Reduction**

Sometimes you may end up with too many topics or too few topics generated, BerTopic gives you an option to control this behavior in different ways. 

```python
new_topics, new_probs = model.reduce_topics(train, topics, probabilities, nr_topics="auto")
```

**Make Prediction**

```python
# convert to list 
test = test.text.to_list()

topics, probs = model.transform(test)
```
